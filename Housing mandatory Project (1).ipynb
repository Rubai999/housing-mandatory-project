# 1. What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.

```python

# step 1-Importing the necessary libraries:

import pandas as pd
import matplotlib.pyplot as plt

# step 2-Reading the Excel file into a pandas DataFrame:

house=pd.read_excel("C:\\Users\\RUBAI PANDEY\\Downloads\\housing.xlsx")

# step 3-Calculating the average median income:

avg_median_income = house['median_income'].mean()
print("Average Median Income:", avg_median_income)

# plt.figure(figsize=(8,4)) sets the figure size to 8 inches wide and 4 inches tall.
plt.figure(figsize=(8,4))

''' plt.hist() creates a histogram using the 'median_income' column from the DataFrame house. 
It specifies bins=20 to divide the range of values into 20 equally spaced bins. 
The edgecolor='black' argument sets the color of the edges of the bars in the histogram.'''

plt.hist(house['median_income'], bins=20, edgecolor='black')


'''plt.xlabel(), plt.ylabel(), and plt.title() are used to set the labels and title for the x-axis, y-axis, and the plot, 
respectively.'''


plt.xlabel('Median Income')
plt.ylabel('Frequency')
plt.title('Distribution of median income')

# plt.show() displays the histogram plot on the screen.
plt.show()
```

```python
'''Here's an explanation of the distribution based on the histogram plot:

1.Shape:

From the histogram, we can observe the shape of the distribution. 
It appears to be slightly right-skewed, meaning that the tail of the distribution extends towards higher median income values. 
This indicates that there are relatively fewer data points with very low median incomes and 
a higher concentration of data points in the middle to higher income range.

2.Peaks and Gaps:

The histogram may exhibit peaks and gaps, indicating concentrations and gaps in the frequency of median income values.
In this case, based on the histogram, we can see a noticeable peak around the range of approximately 3 to 4. 
This suggests that there is a relatively higher frequency of median income values falling within this range.
It implies that a significant number of individuals or households in the dataset have median incomes within this specific range.

3.Range:

The x-axis of the histogram represents the range of median income values divided into bins.
In this case, it seems to span from approximately 0 to 10, indicating the range of median incomes covered by the dataset.

4.Frequency:

The y-axis of the histogram represents the frequency or count of data points falling within each bin.
The height of each bar corresponds to the number of data points (individuals or households)
that have a median income falling within that specific income range.


Overall, based on the provided histogram plot, the distribution of median income values in the dataset appears slightly right-skewed, with a peak around the range of 3 to 4.
This suggests that there is a concentration of individuals or households with median incomes in that range.'''
```

# 6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
def calculate_median(dataset):
    sorted_data = sorted(dataset)
    n = len(sorted_data)
    
    if n % 2 == 0:
        mid1 = n // 2
        mid2 = mid1 - 1
        median = (sorted_data[mid1] + sorted_data[mid2]) / 2
    else:
        mid = n // 2
        median = sorted_data[mid]
    
    return median
```

# 10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python

# step 1-Importing the necessary library:

import pandas as pd

# step 2-Assigning the file path to a variable:
original_dataset_path ="C:\\Users\\RUBAI PANDEY\\Downloads\\housing.xlsx"

# step 3-Reading the Excel file into a pandas DataFrame:
original_dataset = pd.read_excel(original_dataset_path)

# step 4-Creating a new column based on conditions:
original_dataset['total_bedroom_size'] = original_dataset['total_bedrooms'].apply(lambda x: 'small' if x <= 10 else 'medium' if x < 1000 else 'large')

# step 5-printing the result

print(original_dataset.head(5))
```

```python
'''explanation:-
The code essentially reads the original dataset from the Excel file, 
creates a new column 'total_bedroom_size' based on conditions applied to the 'total_bedrooms' column.

we can see that from the result that:--
1.If the value is less than or equal to 10, it is categorized as 'small'.
2.If the value is greater than 10 and less than 1000, it is categorized as 'medium'.
3.If the value is 1000 or greater, it is categorized as 'large'.'''  
```
